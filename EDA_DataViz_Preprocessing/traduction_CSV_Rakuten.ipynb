```python
import pandas as pd
import numpy as np
import time
from googletrans import Translator

#Les quotas quotidiens sont réinitialisés à minuit (heure du Pacifique). 
#Un délai de 24 heures peut donc s'écouler avant que les nouveaux quotas soient appliqués.

#first time only
#data_train = pd.read_csv('X_train_rakuten_update.csv',index_col=0)
#data_train = pd.read_csv('X_train_rakuten_update_with_lang_ind.csv',index_col=0)
#when process in in progress 
data_train = pd.read_csv('X_train_rakuten_update_with_translated_designation.csv',index_col=0)
#column creation first time only !!!!!!!!
#data_train['translated_designation'] = ''
display(data_train.head(2))
translator = Translator(service_urls=['translate.google.com'])
#toto = translator.translate(data_train['designation'].iloc[0],src=data_train['lang_designation'].iloc[0],dest='fr')
#print(toto.text)

#language detection of product designation text
#from googletrans import Translator
#translator = Translator(service_urls=['translate.google.com'])
#column creation first time only !!!!!!!!
#data_train['lang_designation'] = ''
#analysis of language run by batch of 100 samples every 2 mn to be under Google authorized free usage.
# as only 20% designation field has to be translated (not french)so around 600 texts per hour
# control of size limit (5000 characters per text)
# print of oversized text to be reduced manually before elementary translation at end
step = 0
for j in range(1):
    for i in range(step+(100*j),step+(100*(j+1))):
        if (data_train['lang_designation'].iloc[i] == 'fr'):
            data_train['translated_designation'].iloc[i] = data_train['designation'].iloc[i]
        else:
            if (len(data_train['designation'].iloc[i]) >= 5000):
                print("designation supérieur à 5000 caractères :",i)
            else:
                data_train['translated_designation'].iloc[i] = translator.translate(data_train['designation'].iloc[i],src=data_train['lang_designation'].iloc[i],dest='fr').text
    print("loop {0} done".format(j))
    time.sleep(120)

print("fini")
data_train.to_csv(r'X_train_rakuten_update_with_translated_designation.csv',index=True)
```
